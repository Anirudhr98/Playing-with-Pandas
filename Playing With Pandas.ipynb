```python
#Importing Pandas, Reading Csv and displaying the first 5 row data
import pandas as pd
df = pd.read_csv('pokemon_data.csv')
df.head()
```

```python
df.columns
print
```

```python
# To read particular columns
print(df[['Name', 'Type 1', 'HP']])
```

```python
#suppose you want the value of 2nd and 3rd row (Remember indexes start from 0)
print(df.iloc[2][3])
```

```python
df.sort_values(['Attack'], ascending = 'True')
```

```python
#Adding Columns
df['Yolo'] = df['Attack'] + df['HP']
df.head()
```

```python
#Dropping Columns
df.drop(columns = ['Yolo'])
```

```python
#Rearrange Columns
#Create column for Agg HP....Speed = Total which should be displayed after Speed
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed'] 
df.head()
```

```python
cols=list(df.columns.values)
cols
```

```python
df = df[cols[0:9] + [cols[-1]] + cols[9:12]]
df.head()
```

```python
#Filtering Data
#based on grass
df.loc[df['Type 1'] == 'Grass']
```

```python
#based on generation 1 
df.loc[df['Generation'] == 1]
```

```python
#based on generation1 and type of grass
df.loc[(df['Generation'] == 1) & (df['Type 1'] == 'Grass')]
```

```python
#either grass or type 1 pokemon 
df.loc[(df['Generation'] == 1) | (df['Type 1'] == 'Grass')]
```

```python
#STATISTICS 
df.groupby(['Type 1']).mean()
```

```python
#groupby type 1 and find mean . Also sort values by HP


df.groupby(['Type 1']).mean().sort_values('HP')
```

```python
df.groupby(['Type 1']).count().sort_values(['Name'])
```

```python
#Regex Filtering
#Type 1 contains ras
df.loc[df['Type 1'].str.contains('ras')]
```
